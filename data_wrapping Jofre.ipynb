ENRIQUECIMIENTO

En primer lugar, para cada columna que es un array de string generaremos un campo que muestra la cantidad del array.<br>
por ejemplo en la columna genero si el valor es "Action & Adventure, Animation, Crime" el valor en la columna cantidad de generos sera 3 

```python
columnas_array=["Generos","Idioma_original","Creador","Idiomas","Canales","Idiomas_hablados","Companias_Produccion","Paises_Produccion"]
def contar(col):
    if col=='sin informacion':
        return 0
    else:
        return col.count(',')+1
    
for c in columnas_array:
    nombre_col="num_"+c
    tmdb[nombre_col]=tmdb[c].apply(contar)
```

```python
tmdb.head()
```

A continuacion se expandira la informacion de las fechas 

```python
tmdb['Fecha_Inicio']= pd.to_datetime(tmdb['Fecha_Inicio'])
tmdb['Mes_I']= tmdb['Fecha_Inicio'].dt.month
tmdb['estacion_I']=tmdb['Mes_I']%4
tmdb['estacion_I_n']= tmdb['estacion_I'].map({0:'primavera',1:'verano',2:'otoño',3:'invierno'})
tmdb['anho_I']= tmdb['Fecha_Inicio'].dt.year

tmdb['Fecha_Fin']= pd.to_datetime(tmdb['Fecha_Fin'])
tmdb['Mes_F']= tmdb['Fecha_Fin'].dt.month
tmdb['estacion_F']=tmdb['Mes_F']%4
tmdb['estacion_F_n']= tmdb['estacion_F'].map({0:'primavera',1:'verano',2:'otoño',3:'invierno'})
tmdb['anho_F']= tmdb['Fecha_Fin'].dt.year
```

Por ultimo se crearan tablas auxiliares asociadas a las columnas de array

```python

def expand_array(df,column):

    df_aux=df[[column]].copy()
    df_aux["id"]=df_aux.index


    new_df=pd.DataFrame({'id_show' : [],column:[]})
    for index, row in df_aux.iterrows():
        aux_id=row['id']
        aux_column=row[column]
        list_column=aux_column.split(",")
        for i in list_column:
            new_row=pd.DataFrame([{'id_show':aux_id,column:i}])
            new_df=pd.concat([new_df,new_row],ignore_index=True)
    return new_df

def create_csv(df,column):
    name=column+".csv"
    df.to_csv(name,index=False)

```

```python
columnas_array=["Generos","Idioma_original","Creador","Idiomas","Canales","Idiomas_hablados","Companias_Produccion","Paises_Produccion"]

df_columnas={}

for c in columnas_array:
    newdf=expand_array(tmdb,c)
    df_columnas[c]=newdf
```

```python
x=0
for d in df_columnas:
    print(d)
    column =columnas_array[x]
    create_csv(df_columnas[d],column)
    x+=1
```
